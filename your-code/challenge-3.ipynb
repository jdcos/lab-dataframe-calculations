## How to label customers whose aggregated `amount_spent` is in a given quantile range?


We break down the main problem into several sub problems:

#### Sub Problem 1: How to aggregate the  `amount_spent` for unique customers?

#### Sub Problem 2: How to select customers whose aggregated `amount_spent` is in a given quantile range?

#### Sub Problem 3: How to label selected customers as "VIP" or "Preferred"?

*Note: If you want to break down the main problem in a different way, please feel free to revise the sub problems above.*

Now in the workspace below, tackle each of the sub problems using the iterative problem solving workflow. Insert cells as necessary to write your codes and explain your steps.

```python
# 1.1
customer_spending = orders.groupby('CustomerID')['amount_spent'].sum().reset_index()
customer_spending
```

```python
# 1.2
vip_threshold = customer_spending['amount_spent'].quantile(0.90)
customer_spending[customer_spending['amount_spent'] >= vip_threshold]
```

```python
preferred_threshold = customer_spending['amount_spent'].quantile(0.75)
customer_spending[customer_spending['amount_spent'] >= preferred_threshold]
```

```python
# 1.3
def label_customer(row):
    if row['amount_spent'] >= vip_threshold:
        return 'VIP'
    elif row['amount_spent'] >= preferred_threshold:
        return 'Preferred'
    else:
        return 'Regular'

customer_spending['CustomerLabel'] = customer_spending.apply(label_customer, axis=1)
```

## Q2: How to identify which country has the most VIP Customers?

```python
customer_spending_vip = orders.merge(customer_spending, on="CustomerID")
customer_spending_vip
```

```python
customer_spending['CustomerLabel'] = customer_spending['amount_spent'].apply(
    lambda x: 'VIP' if x >= vip_threshold else 'Regular')
```

```python
customer_data = orders[['CustomerID', 'Country']].drop_duplicates()
vip_customers = customer_spending[customer_spending['CustomerLabel'] == 'VIP']
vip_customers_with_country = vip_customers.merge(customer_data, on='CustomerID')
```

```python
vip_counts_by_country = vip_customers_with_country['Country'].value_counts().reset_index()
vip_counts_by_country.columns = ['Country', 'VIP_Count']
```

```python
vip_counts_by_country
```

## Q3: How to identify which country has the most VIP+Preferred Customers combined?

```python
customer_spending['CustomerLabel'] = customer_spending['amount_spent'].apply(label_customer)

# Step 4: Merge with country information
customer_data = orders[['CustomerID', 'Country']].drop_duplicates()
labeled_customers = customer_spending.merge(customer_data, on='CustomerID')
```

```python
# Step 5: Count VIP and Preferred customers by country
vip_preferred_counts_by_country = labeled_customers[labeled_customers['CustomerLabel'].isin(['VIP', 'Preferred'])]
country_counts = vip_preferred_counts_by_country['Country'].value_counts().reset_index()
country_counts
```

```python
country_counts.columns = ['Country', 'VIP_Preferred_Count']
country_counts
```

```python
# Display the country with the most VIP + Preferred customers
top_country = country_counts.iloc[0]
print("Country with the most combined VIP and Preferred customers:", top_country['Country'], "with", top_country['VIP_Preferred_Count'], "VIP + Preferred customers.")
```
